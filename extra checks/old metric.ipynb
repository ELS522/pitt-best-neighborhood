# Which Pittsburgh neighbourhood has the most parks?
Parks create a healthy and enjoyable community.  
In this project, we use data from the **Western Pennsylvania Regional Data Center (WPRDC)** about parks maintained or tracked by the **City of Pittsburgh**.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
parks = pd.read_csv("parks.csv")
parks.head()
```

```python
parks = parks.dropna(subset=['neighborhood'])

parks_per_neighborhood = parks["neighborhood"].value_counts()

parks_per_neighborhood.head(10)
```

```python
plt.figure(figsize=(10,6))
parks_per_neighborhood.head(10).plot(kind='bar', color='mediumseagreen')
plt.title("Top 10 Pittsburgh Neighborhoods by Number of Parks")
plt.xlabel("Neighborhood")
plt.ylabel("Number of Parks")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
