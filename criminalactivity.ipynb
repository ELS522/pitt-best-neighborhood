# Which City In Pittsburgh has the least amount of criminal activiy


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crime_df = pd.read_csv('criminalacitivity.csv', low_memory=False)
print('Preview of dataset:')
display(crime_df.head())
crime_df = crime_df.dropna(subset=['Neighborhood'])
```

```python
crime_counts = crime_df['Neighborhood'].value_counts().reset_index()
crime_counts.columns = ['Neighborhood', 'CrimeCount']
```

```python
safest = crime_counts.iloc[-1] # neighborhood with fewest crimes
print(f"Safest neighborhood: {safest['Neighborhood']} with {safest['CrimeCount']} reported crimes.")
```

```python
plt.figure(figsize=(12,6))
plt.bar(crime_counts['Neighborhood'], crime_counts['CrimeCount'], color='skyblue')
plt.xticks(rotation=90)
plt.title('Total Reported Crimes by Neighborhood in Pittsburgh')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Reported Crimes')
plt.tight_layout()
plt.show()
```

```python
print("\n### Analysis Narrative")
print("Based on the number of reported crimes, the safest neighborhood in Pittsburgh is", safest['Neighborhood'], ".")
```
