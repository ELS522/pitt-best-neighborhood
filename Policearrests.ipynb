# Which City In Pittsburgh has the least Police arrests


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

arrests_df = pd.read_csv('policearrest.csv')
print('Preview of dataset:')
display(arrests_df.head())
print("Total Police arrests:", len(arrests_df))
```

In the original dataset all crime was included, of course, all crime is not exactly violent.
I'm sure most would rather hear about shoplifting arrests than murders and homocides.
In order to accommadate, we decided to only evaluate violent crimes which include:
Assault Offenses, robbery, Homocide offenses, Kidnapping/Abduction, and sex offense.

```python
arrests_df = arrests_df.dropna(subset=['Neighborhood'])
# Define violent crime categories
violent_crimes = [
    "Assault Offenses",
    "Robbery",
    "Homicide Offenses",
    "Kidnapping/Abduction",
    "Sex Offenses"
]

# Filter dataset for violent crimes only
violent_df = arrests_df[arrests_df['NIBRS_Offense_Category'].isin(violent_crimes)]

# Check the size of this filtered dataset
print("Total violent crime arrests:", len(violent_df))
violent_df.head()
```

```python
# Count violent arrests by neighborhood
violent_counts = violent_df['Neighborhood'].value_counts().reset_index()
violent_counts.columns = ['Neighborhood', 'Violent_Arrest_Count']

# Sort from fewest to most arrests
violent_counts = violent_counts.sort_values(by='Violent_Arrest_Count', ascending=True)

# Display the 10 neighborhoods with the fewest violent crime arrests
violent_counts.head(10)
```

```python
most_peaceful = violent_counts.iloc[0]

print(f" The most peaceful neighborhood in Pittsburgh is **{most_peaceful['Neighborhood']}**, "
      f"with only {most_peaceful['Violent_Arrest_Count']} violent crime arrests recorded.")
```

```python
# Select the 10 neighborhoods with the fewest violent arrests
bottom_10 = violent_counts.head(10)

# Create a bar chart
plt.figure(figsize=(10,6))
bars = plt.barh(bottom_10['Neighborhood'], bottom_10['Violent_Arrest_Count'], color='mediumseagreen')

# Add chart details
plt.title('🕊️ 10 Most Peaceful Neighborhoods in Pittsburgh (Fewest Violent Arrests)', fontsize=14, fontweight='bold')
plt.xlabel('Number of Violent Arrests')
plt.ylabel('Neighborhood')

# Label each bar with its value
for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
             f"{int(bar.get_width())}", va='center')

plt.tight_layout()
plt.show()
```
